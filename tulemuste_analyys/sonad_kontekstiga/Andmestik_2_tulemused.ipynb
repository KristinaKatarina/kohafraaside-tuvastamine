```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('500_sona_tulemused.csv')
```

```python
# arvutame protsendi, mitmes lauses antud sõna oli koht
for index, row in df.iterrows():
    on_koht = int(row["on_loc"])
    lauseid = int(row["esinemise_korrad"])
    protsent = (on_koht * 100) / lauseid
    df.at[index, "protsent"] = protsent
```

```python
df['protsent_rounded'] = df['protsent'].round(1)  # round to 1 decimal
value_counts = df['protsent_rounded'].value_counts().sort_index()

plt.figure(figsize=(12, 6))
plt.bar(value_counts.index.astype(str), value_counts.values, color='gray', edgecolor='black')
plt.xlabel('Protsentuaalne vastavus (%)')
plt.ylabel('Sõnade arv')
plt.title('Sõnade vastavus füüsilisele kohale lausetes')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('500_sõna.png', format='png')
plt.show()
```
