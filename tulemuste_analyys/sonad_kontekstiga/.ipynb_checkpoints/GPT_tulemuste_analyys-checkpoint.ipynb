```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('GPT4_tulemused_kontekstiga.csv')
df = df.rename(columns={
    "belongings": "ese",
    "clothing_items": "riideese",
    "edibles": "jook_toit",
    "enclosures_and_vehicles": "saab_sisse_minna",
    "geographical_locations": "geograafiline_lokatsioon",
    "institutions": "asutus",
    "nonsense_words": "tähenduseta_sõna",
    "organisations_and_events": "organisatsioon_üritus_firma",  
    "persons": "isikunimi",
    "platforms_and_floors": "saab_peal_seista",
    "spatial_directions": "suund",
    "spatial_region": "piirkond",
    "time_expressions": "ajaperiood",
    "vegetative_enclosure": "taim"
})

columns = df.columns.difference(['lemma'])
```

```python
plt.figure(figsize=(4 * 3, 3 * 5))
for i, column in enumerate(columns):
    plt.subplot(5, 3, i + 1)
    df_clean = df.dropna(subset=[column])
    value_counts = df_clean[column].value_counts().sort_index()
    
    all_values = pd.Series(index=range(6), dtype=int)
    value_counts = all_values.add(value_counts, fill_value=0)
    
    plt.bar(value_counts.index, value_counts.values, color='gray', edgecolor='black')
    plt.xlabel('Alamklassile vastavate lausete arv')
    plt.ylabel('Esinemiste arv')
    plt.title(column)
    plt.xticks(range(6)) 
    plt.grid(axis='y', linestyle='--', alpha=0.7)


plt.tight_layout()
#plt.savefig('histogramm.png', format='png')
plt.show()
```

```python
# lisa joonis heatmapina, et näha sõnade tähenduse muutumist olenevalt lausest
import seaborn as sns
import numpy as np
from matplotlib.colors import TwoSlopeNorm

heatmap_data = df.set_index('lemma')


cmap = sns.diverging_palette(0, 0, as_cmap=True)  # punane → valge → punane
norm = TwoSlopeNorm(vmin=0, vcenter=2.5, vmax=5)

plt.figure(figsize=(12, 8))
sns.heatmap(
    heatmap_data,
    cmap=cmap,
    norm=norm,
    linewidths=0.5,
    linecolor='gray',
    cbar_kws={'label': 'Esinemissagedus'},
    yticklabels=False  
)

plt.title('Semantiliste alamklasside esinemissageduse soojuskaart')
plt.xlabel('Semantilised alamklassid')
plt.ylabel('Sõnad')
plt.tight_layout()
plt.show()
```
