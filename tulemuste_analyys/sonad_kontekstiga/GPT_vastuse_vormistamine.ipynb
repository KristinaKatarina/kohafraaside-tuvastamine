```python
import pandas as pd
import re

# leiame 'response' veerust kõik sõnade ja väärtuste paarid
def extract_pairs(text):
    if not isinstance(text, str):
        return []
    
    word = re.findall(r'"word": "([^"]*)"', text) #leiab järgmisel kujul sõned ["word": "naftariigil"]
    märgendused = re.findall(r'"(?:\w|\s)+":\s*"?(0|1)"?(?!\d)', text) #leiab järgmisel kujul sõned ["evaluation": "0"]
    if len(märgendused) != 5 or len(word) < 1:
        print(text)
    else:
        value = 0
        for märgendus in märgendused:
            value += int(märgendus)
    word = word[0]   
    # parandame encoding vigu
    word = re.sub(r'(\Wu00e4)', 'ä', word)
    word = re.sub(r'(\Wu00f5)', 'õ', word)
    word = re.sub(r'(\Wu00fc)', 'ü', word)
    word = re.sub(r'(\Wu00f6)', 'ö', word)
    word = re.sub(r'(\Wu00dc)', 'Ü', word)
    
    return word, int(value)
```

```python
# loome uue dataframe'i, kus on iga leitud sõna kohta rida
# igas reas on sõna, semantiline_klass ja ennustus

rows = []
i = 0
# Loop through each row in the original DataFrame
for _, row in df.iterrows():
    i+=1
    semantic_class = row['semantic_class']
    word_value_pairs = extract_pairs(row['response'])
    rows.append({
            'lemma': word_value_pairs[0],
            'väärtus': int(word_value_pairs[1]),
            'semantiline_klass': semantic_class
        })

simple_df = pd.DataFrame(rows)
simple_df
```

```python
# Saame loodud dataframe'ist lõpliku tulemuse, kus on semantilised_klassid veergude nimed
final_df = simple_df.pivot_table(index='lemma', 
                            columns='semantiline_klass', 
                            values='väärtus', 
                            aggfunc='first') 




final_df.columns.name = None 
final_df = final_df.reset_index()
final_df
```
