```python
import pandas as pd

faili_nimi = 'dataset_500_gpt4o_annotation.csv'

df = pd.read_csv(faili_nimi, delimiter='|')
```

```python
df
```

```python
df = df.rename(columns={"lemma": "sõna"})

# Group and calculate
result_df = df.groupby("sõna").agg(
    on_loc=("gpt_annotation", lambda x: (x == "LOC").sum()),
    esinemise_korrad=("sõna", "count")
).reset_index()

# Cap the count at 5
result_df["esinemise_korrad"] = result_df["esinemise_korrad"].clip(upper=5)

# Preview
print(result_df)
```

```python
result_df.to_csv('500_sona_tulemused.csv', index=False)
```
