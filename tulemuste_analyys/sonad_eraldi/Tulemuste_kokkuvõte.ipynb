```python
import pandas as pd
```

```python
# ChatGPT täpsus puhtalt 0 ja 1 võrreldes

# tõlgime minu märgenduse inglise keelde
df_minu = df_minu.rename(columns={
    "ese": "belongings",
    "riide_ese": "clothing_items",
    "jook_toit": "edibles",
    "sisse_minna": "enclosures_and_vehicles",
    "geo_lok": "geographical_locations",
    "asutus": "institutions",
    "-": "nonsense_words",
    "org_ür_fir": "organisations_and_events",  
    "isikunimi": "persons",
    "peal_seista": "platforms_and_floors",
    "suund": "spatial_directions",
    "piirkond": "spatial_region",
    "ajaperiood": "time_expressions",
    "taim": "vegetative_enclosure"
})
df_minu
```

```python
def kontrolliRida(df_gpt, kontrollitav_lemma, rida):
    # 
    veerud = [
    "belongings",
    "clothing_items",
    "edibles",
    "enclosures_and_vehicles",
    "geographical_locations",
    "institutions",
    "nonsense_words",
    "organisations_and_events",
    "persons",
    "platforms_and_floors",
    "spatial_directions",
    "spatial_region",
    "time_expressions",
    "vegetative_enclosure"
]

    leitud_lemmad = df_gpt[df_gpt['lemma'] == kontrollitav_lemma]
    if not leitud_lemmad.empty:
        sama = 0
        erinev = 0
        leitud_rida = leitud_lemmad.iloc[0]
        for veeru_nimi in veerud:
            if (leitud_rida[veeru_nimi] == rida[veeru_nimi]):
                sama +=1
            elif not pd.isna(leitud_rida[veeru_nimi]):  
                erinev +=1
        return [sama, erinev]       
            
    else:
        return [0,0]
```

```python
def kontrolliTervetRida(df_gpt, kontrollitav_lemma, rida):
    # 
    veerud = [
    "belongings",
    "clothing_items",
    "edibles",
    "enclosures_and_vehicles",
    "geographical_locations",
    "institutions",
    "nonsense_words",
    "organisations_and_events",
    "persons",
    "platforms_and_floors",
    "spatial_directions",
    "spatial_region",
    "time_expressions",
    "vegetative_enclosure"
]

    leitud_lemmad = df_gpt[df_gpt['lemma'] == kontrollitav_lemma]
    if not leitud_lemmad.empty:
        leitud_rida = leitud_lemmad.iloc[0]
        for veeru_nimi in veerud:
            if (leitud_rida[veeru_nimi] != rida[veeru_nimi]):
                return 0
        return 1  
            
    else:
        return 0
```

```python
sama_35 = 0
erinev_35 = 0
sama_4 = 0
erinev_4 = 0

kokku = 0
sama_rida_35 = 0
sama_rida_4 = 0
for index, row in df_minu.iterrows():
    kontrollitav_lemma = row["lemma"]
    # leiame sõne gpt 3.5 tulemustest
    tulemus_35 = kontrolliRida(df_GPT35, kontrollitav_lemma,row)
    sama_35 += tulemus_35[0]
    erinev_35 += tulemus_35[1]

    # leiame sõne gpt 4o tulemustest
    tulemus_4 = kontrolliRida(df_GPT4, kontrollitav_lemma,row)
    sama_4 += tulemus_4[0]
    erinev_4 += tulemus_4[1]

    # täpsus terve rea kaupa
    sama_rida_35 += kontrolliTervetRida(df_GPT35, kontrollitav_lemma,row)
    sama_rida_4 += kontrolliTervetRida(df_GPT4, kontrollitav_lemma,row)
    kokku +=1

täpsus_35 = sama_35/(sama_35+erinev_35)    
print( "GPT3.5: " + str(täpsus_35))

täpsus_4 = sama_4/(sama_4+erinev_4)    
print("GPT4o: " + str(täpsus_4))

täpsus_35_terve = sama_rida_35/(kokku)    
print("GPT3.5 terve rida täpsus: " + str(täpsus_35_terve))

täpsus_4_terve = sama_rida_4/(kokku)    
print("GPT4o terve rida täpsus: " + str(täpsus_4_terve))
```

```python
# täpsuse arvutamine kui võrrelda, kas on koht või ei ole

def kohaMääramine(df_gpt):
    df_gpt['0'] = 0
    df_gpt['1'] = 0
    df_gpt['?'] = 0
    df_gpt['-'] = 0

    koha_veerud = [
        "belongings", "enclosures_and_vehicles", "geographical_locations",
        "platforms_and_floors", "spatial_directions", "spatial_region",
        "vegetative_enclosure"
    ]
    mitte_koha_veerud = [
        "clothing_items", "edibles", "persons",
        "time_expressions", "organisations_and_events"
    ]

    for index, row in df_gpt.iterrows():
        if row['nonsense_words'] == 1 and row['geographical_locations'] == 0 and row['persons'] == 0:
            df_gpt.loc[index, '-'] = 1
            continue
        if row['institutions'] == 1:
            df_gpt.loc[index, '?'] = 1
            continue

        onKoht = any(row[veerg] == 1 for veerg in koha_veerud)
        eiOleKoht = any(row[veerg] == 1 for veerg in mitte_koha_veerud)

        if not onKoht:
            df_gpt.loc[index, '0'] = 1
        elif eiOleKoht:
            df_gpt.loc[index, '?'] = 1
        else:
            df_gpt.loc[index, '1'] = 1
```

```python
# määrame GPT tulemuste põhjal kas sõnad on kohad või mitte
kohaMääramine(df_GPT35)
kohaMääramine(df_GPT4)
df_minu = pd.read_csv('minu_tulemused.csv')
```

```python
def kontrolliKasOnKoht(df_gpt, kontrollitav_lemma, õige_veerg):
    leitud_lemmad = df_gpt[df_gpt['lemma'] == kontrollitav_lemma]
    if not leitud_lemmad.empty:
        leitud_rida = leitud_lemmad.iloc[0]
        if (leitud_rida[õige_veerg] == 1):
                return 1
    return 0    
```

```python
kokku = 0
õigeid_35 = 0
õigeid_4 = 0


for index, row in df_minu.iterrows():
    kontrollitav_lemma = row["lemma"]
    koha_veerud = ['0', '1', '-', '?']
    õige_veerg = row[koha_veerud].idxmax()
    õigeid_35 += kontrolliKasOnKoht(df_GPT35, kontrollitav_lemma, õige_veerg)
    õigeid_4 += kontrolliKasOnKoht(df_GPT4, kontrollitav_lemma, õige_veerg)
    kokku += 1

täpsus_35 = õigeid_35/kokku
täpsus_4 = õigeid_4/kokku

print("GPT3.5 koha täpsus: " + str(täpsus_35))
print("GPT4o koha täpsus: " + str(täpsus_4))
```
