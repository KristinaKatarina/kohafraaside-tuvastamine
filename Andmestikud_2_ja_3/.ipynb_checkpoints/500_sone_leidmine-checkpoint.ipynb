```python
import pandas as pd

df = pd.read_csv('soned_1000_algne.csv')

# jätame alles esimesed 500 lemmat
lemma_list = df['lemma'].head(500).tolist()

```

```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('BT_andmed.db')  

df_suur = pd.read_sql_query("SELECT * FROM 'table'", conn) 
conn.close()
```

```python
filtered_df = df_suur[df_suur['lemma'].isin(lemma_list)] # jätame alles ainult andmestik 1 sõnad
```

```python
filtered_df
```

```python
# leiame sõnad, mida andmetes ei esine
found_lemmas = set(filtered_df['lemma'].unique())
all_lemmas = set(lemma_list)
missing_lemmas = all_lemmas - found_lemmas
missing_lemmas = sorted(missing_lemmas)
print(f"{len(missing_lemmas)} lemmas had no matches:")
print(missing_lemmas)
```
